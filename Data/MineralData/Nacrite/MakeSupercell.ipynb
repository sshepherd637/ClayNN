```python
from ase.io import read, write
from ase import build
from ase.visualize import view
import numpy as np
```

```python
nacrite = read('Nacrite.xyz')
view(nacrite, viewer='x3d')
```

```python
build_cell = np.array([[2,0,0],[0,2,0],[0,0,1]])
new_nacrite = build.make_supercell(nacrite, build_cell)
```

```python
view(new_nacrite, viewer='x3d')
```

```python
write('Nacrite.2x2x1.xyz', new_nacrite)
```
