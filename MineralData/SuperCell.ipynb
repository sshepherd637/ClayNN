```python
from ase.io import read, write
from ase.visualize import view
import ase.build as build
from ase import Atoms

import numpy as np
```

```python
frames = read('nacrite.xyz')
#frames.wrap(eps=1e-2)
view(frames, viewer='nglview')
```

```python
transform = np.array([[2,0,0],
                      [0,2,0],
                      [0,0,2]])

newframes = build.make_supercell(frames, transform)
view(newframes, viewer='nglview')
z_coords = []
for atom in newframes:
    if atom.symbol == 'O':
        #print(atom.position[2])
        z_coords.append(atom.position[2])

z_coords.sort()
print(z_coords)
        
range = (3.1782235088, 17.7507011412)
print(newframes.cell)
#double_layer = 
new_atoms = []
for ii, atom in enumerate(newframes):
    #print(atom.position)
    #print(atom.position[2])
    if atom.position[2] >= range[0] and atom.position[2] < range[1]:
        #print(ii)
        new_atoms.append(atom)

        
new_height = range[1] - range[0]

len(new_atoms)
new_layers = Atoms(new_atoms, 
                   cell=[[17.82, 0.0, 0.0], [0.0, 10.288, 0.0], [-5.318726047, 0.0, new_height]],
                   pbc=[1,1,1])

for atom in new_layers:
    atom.position[2] -= 3.1782235088
new_layers.wrap(eps=1e-2)
```

```python
new_layers.wrap(eps=1e-30)
view(new_layers, viewer='nglview')
```

```python
write('dickite.2x2x1.xyz', newframes)
```
